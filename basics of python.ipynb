```python
n=10
i=0

while i<n:
    print(i)
    i=i+1
    if i==5:
        break

else:
    print('while is sucessfully executed')

```

```python
# continue
n=10
i=0

while i<n:
    print(i)
    i=i+1
    if i==5:
        continue

else:
    print('while is sucessfully executed')

```

```python
print('my name is {fname} and age is {fage}'.format(fname=name,fage=age))
```

```python
name='pooja singh'
age=23
degree='btech'
print('my age is {} and i have completed : my degree of {} with the name of{} '.format(age,degree,name))
```

```python
# with parameter
name='pooja singh'
age=23
degree='btech'
print('my age is {fage} and i have completed : my degree of {fdegree} with the name of{fname} '.format(fage=age,fdegree=degree,fname=name))
```
